```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import f1_score
from sklearn.metrics.pairwise import cosine_similarity
from statistics import harmonic_mean
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import HashingVectorizer
```

```python
df = pd.read_csv('Coursera.csv')
df.head()
```

```python
df.drop_duplicates(inplace=True)

#drop Coursera Project Network rows in University
df = df[df['University'] != 'Coursera Project Network']

#drop rows with Not Calibrated in Course Rating
df = df[df['Course Rating'] != 'Not Calibrated']

```

```python
#df.drop('Unnamed: 0',axis=1,inplace=True)
print(df.isna().sum())
df.head()
```

```python
df.to_csv('CourseraData.csv')
```

```python
vectorizer = TfidfVectorizer(stop_words='english')
vectors = vectorizer.fit_transform(df['Course Name'])
```

```python
def recommend_by_course_title (title, recomm_count=10) : 
    title_vector = vectorizer.transform([title])
    cosine_sim = cosine_similarity(vectors, title_vector)
    idx = np.argsort(np.array(cosine_sim[:,0]))[-recomm_count:]
    sdf = df.iloc[idx].sort_values(by='Course Rating', ascending=False)
    return sdf
```

```python
recommend_by_course_title('Machine Learning')
```
